```python
predictions = ["एमएनपीए शिक्षक संघ के अध्यक्ष राजेश गावरे ने पुरस्कार देकर विद्यालय को सम्मानित किया।",
               "रमेश सातपुते ने किले का निरीक्षण किया।",
               "किले के निर्माण में निखिल कावले, दर्शन गेडेकर, साहिल मेश्राम जैसे छात्रों ने भाग लिया।"]
references = ["मनपा शिक्षक संघ के अध्यक्ष राजेश गवरे ने स्कूल को भेंट देकर सराहना की।",
              "किले का परीक्षण रमेश सातपुते ने किया।",
              "किला निर्माण में निखिल कावले, दर्शन गेड़ेकर, साहिल मेश्राम इन विद्यार्थियों ने सहभाग लिया।"]
res = score(cands=predictions, refs=references, lang = "hi")
res[0]
```

```python
predictions = ["एमएनपीए शिक्षक संघ के अध्यक्ष राजेश गावरे ने पुरस्कार देकर विद्यालय को सम्मानित किया।",
               "रमेश सातपुते ने किले का निरीक्षण किया।",
               "किले के निर्माण में निखिल कावले, दर्शन गेडेकर, साहिल मेश्राम जैसे छात्रों ने भाग लिया।"]
references = ["मनपा शिक्षक संघ के अध्यक्ष राजेश गवरे ने स्कूल को भेंट देकर सराहना की।",
              "किले का परीक्षण रमेश सातपुते ने किया।",
              "किला निर्माण में निखिल कावले, दर्शन गेड़ेकर, साहिल मेश्राम इन विद्यार्थियों ने सहभाग लिया।"]
sc = scorer.BERTScorer(lang = "hi")
res = sc.score(cands=predictions, refs=references)
res[0]
```
